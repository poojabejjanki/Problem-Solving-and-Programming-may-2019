#### Problem statement
program to delete a row in csv file

```python
def updaterowcsv(filename,rowindex,rowdata):
    df=pd.read_csv(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)
    return df
rowdata=[0,'10101010','statename',123,123,123,123,123,123,123,134,123445]
updaterowcsv('data files/income.csv',7,rowdata)
```

```python
def deleterowcsv(filename,rowindex,rowdata):
    df=pd.read_csv(filename)
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df
rowdata=[0,'10101010','statename',123,123,123,123,123,123,123,134,123445]
deleterowcsv('data files/income.csv',4,rowdata)
```
